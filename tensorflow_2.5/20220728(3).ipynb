그림의 구조
> 맷프롭릿가 그리음 그림은 figure, axes,axis 객체 등으로 구성된다.

```python
import numpy as np
np.random.seed(0) # seed를 0으로 고정된다. 따라서 랜덤값이 항상 똑같이 나오게 된다.
f1 = plt.figure(figsize=(10,2))
plt.title('figure size:(10,2)')
plt.plot(np.random.randn(100))
plt.show()
```

```python
f2 = plt.figure(1)
plt.title('Current Figure')
plt.plot([1,2,3,4],'ro:')

f3 = plt.gcf()
print(f2,id(f2))
print(f3,id(f3))

plt.show()
```

```python
x1 = np.linspace(0.0,5.0)
x2 = np.linspace(0.0,2.0)
y1 = np.cos(2*np.pi*x1)*np.exp(-x1)
y2 = np.cos(2*np.pi*x2)

ax1 = plt.subplot(2,1,1) # 2행1열 생성 - '1'행은 내가
plt.plot(x1,y1,'yo-')
plt.title('A table of 2 subplots')
plt.ylabel('Damped oscillation')

ax2 = plt.subplot(2,1,2) # 2행 1열 생성 -'2'행은 내가
plt.plot(x2,y2,'r.-')
plt.title('time(s)')
plt.ylabel('Undamped')

plt.tight_layout() # ax1과ax2를 붙인다.
plt.show()
```

```python
fig, axes = plt.subplots(2,2)

np.random.seed(0)

axes[0,0].plot(np.random.rand(5))
axes[0,0].set_title('axes1')

axes[0,1].plot(np.random.rand(5))
axes[0,1].set_title('axes2')

axes[1,0].plot(np.random.rand(5))
axes[1,0].set_title('axes3')

axes[1,1].plot(np.random.rand(5))
axes[1,1].set_title('axes4')

plt.tight_layout()
plt.show()
```

x측 공유 / y축 양옆

```python
fig,ax0 = plt.subplots()

ax1 = ax0.twinx()
ax0.set_title('2 y 1 x 1 figure')
ax0.plot([10,5,2,9,7],'r-',label='y0')
ax0.set_ylabel('y0')
ax0.grid(False)

ax1.plot([100,200,220,180,120],'g:',label='y1')
ax1.set_ylabel('y1')
ax1.grid(False)
ax0.set_xlabel('twin x')

plt.show()
```
